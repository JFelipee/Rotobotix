```python
import cv2
# Python 2.7
# from scipy.misc import imread, imsave
# Para Python > 3.5 o para Python < 3.5 -> pip install imageio
from imageio import imread, imsave
from matplotlib import pyplot as plt
import select_pixels as sel
```

```python
#Captura de imagenes 
capture = cv2.VideoCapture('video2017-3.avi')

while (True):
    # Leo una imagen del video
    ret,im = capture.read()
 
    if not ret:
        print('Se acabo el video.')
        cv2.destroyAllWindows()
        break
        
    # Muestro la imagen en una ventala OpenCV
    cv2.imshow("Captura", im)

    if cv2.waitKey(35) & 0xFF ==ord('w'):
        break


# Cierro la ventana 
cv2.destroyAllWindows() 
capture.release()
```

```python
# Transformo la imagen al formato Python
imNumpy=cv2.cvtColor(im,cv2.COLOR_BGR2RGB)

# Llamo a la funcion para etiquetar la imagen
markImg = sel.select_fg_bg(imNumpy)

plt.imshow(markImg)
plt.show()
```

```python
# Guardo las imagenes
imsave('imgs/imagenOrigenPelota.png',imNumpy)
imsave('imgs/imagenMarcadaPelota.png',markImg)
```
